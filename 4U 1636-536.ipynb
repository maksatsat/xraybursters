# 4U 1636-536

```python
from ui import efsearch
```

```python
name = '4U 1636-536'
bursts_url = 'https://burst.sci.monash.edu/aqoutput?dtype=bursts&fields=name%2Ctime%2Cinstr&output=topcat&timef=mjd&qfield=name&query_op=%3D&query_val=4U+1636-536'
osbervations_url = 'https://burst.sci.monash.edu/aqoutput?dtype=observations&fields=name%2Ctstart%2Cinstr%2Ctstop&output=topcat&timef=mjd&qfield=name&query_op=%3D&query_val=4U+1636-536'
```

```python
num_simulations = 100
bins = 10
n_steps = 1000
hper, hstat, hstat_expo, hxstats, dper, dstat, dstat_expo, dxstats, yper, ystat, ystat_expo, yxstats = efsearch(bursts_url, osbervations_url, bins=bins, n_steps=n_steps, num_simulations=num_simulations)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(20,10))

plt.plot(np.array(hper)/(3600), hxstats[0], 'o-', color='gray', markersize=0.1, linewidth=0.1, label=' ')
for i in range(1, num_simulations):
    plt.plot(np.array(hper)/(3600), hxstats[i], 'o-', color='gray', markersize=0.1, linewidth=0.1)

plt.plot(np.array(hper)/(3600), hstat_expo, 'bo-', markersize = 1.5, label=name + '   ')
plt.plot(np.array(hper)/(3600), hstat, 'ro-', markersize = 1.5, label=name + '   ')
# plt.plot(np.array(per)/(365*24*3600), stat2, 'ro', markersize = 1.5)
fontsize = 20
labelsize = 20
plt.xlabel('Period, hours', fontsize=fontsize)
plt.ylabel('$\chi^2$', fontsize=fontsize)
plt.grid()
plt.yscale('log') 
# plt.xlim(10,15)
# plt.title(name + '\n   ', fontsize=labelsize)
plt.title(name, fontsize=labelsize)
plt.xticks(fontsize=labelsize)
plt.yticks(fontsize=labelsize)
plt.legend(fontsize=labelsize, loc = 1)

# plt.annotate('max: 12.77 years', (12.77, 557), xytext=(6, 500), arrowprops=dict(color='gray', width=0.3), fontsize=15)
# plt.xlim(150,200)
plt.savefig('results/' + name +'/hours.eps')
plt.savefig('results/' + name +'/hours.png')
plt.show()
```
